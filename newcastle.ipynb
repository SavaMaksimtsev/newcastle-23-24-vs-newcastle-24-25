```python
import matplotlib.pyplot as plt
import numpy as np
from pyfonts import load_font

font_regular = load_font(
    font_url="https://github.com/SavaMaksimtsev/raphinha/raw/main/GoogleSans-Regular.ttf"
)
font_bold = load_font(
    font_url="https://github.com/SavaMaksimtsev/raphinha/raw/main/GoogleSans-Bold.ttf"
)

metrics = ['  ', '   ', 
              '    ', 'xGA  90 ', '   ']
values_2023_24 = [46.66, 46.11, 79.95, 1.51, 5.52]
values_2024_25 = [42.72, 46.89, 82.44, 1.84, 2.94]

fig, ax = plt.subplots(figsize=(8, 8))
bar_width = 0.35
fig.patch.set_facecolor('#181818')
ax.patch.set_facecolor('#181818')

bars_2023_24 = ax.bar(np.arange(len(metrics)) - bar_width/2, values_2023_24, bar_width, label='2023-24', color='#e05451')
bars_2024_25 = ax.bar(np.arange(len(metrics)) + bar_width/2, values_2024_25, bar_width, label='2024-25', color='#3a83c4')

ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['right'].set_color('#181818')
ax.spines['top'].set_color('#181818')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

ax.set_xticks(np.arange(len(metrics)))
ax.set_xticklabels(metrics, rotation=45, ha='right', color='white')

ax.set_title('-23/24 vs -24/25', color='white', font=font_bold, fontsize=18)
ax.legend()

for bar in bars_2023_24 + bars_2024_25:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom', color='white', font=font_regular)

plt.tight_layout()
plt.savefig("newcastle.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['  ', '   ', 
              '    ', 'xGA  90 ', '   ']
values_2023_24 = [46.66, 46.11, 79.95, 1.51, 5.52]
values_2024_25 = [42.72, 46.89, 82.44, 1.84, 2.94]

# Create figure and axis
fig, ax = plt.subplots()

# Set background color to black
fig.patch.set_facecolor('black')
ax.patch.set_facecolor('black')

# Create bars
bar_width = 0.35
x = np.arange(len(categories))
ax.bar(x - bar_width/2, values_2023_24, bar_width, label='2023-24', color='white')
ax.bar(x + bar_width/2, values_2024_25, bar_width, label='2024-25', color='white')

# Add values on top of the columns
for i, value in enumerate(values_2023_24):
    ax.text(i - bar_width/2, value + 0.5, f'{value:.2f}', ha='center', va='bottom', color='white')
for i, value in enumerate(values_2024_25):
    ax.text(i + bar_width/2, value + 0.5, f'{value:.2f}', ha='center', va='bottom', color='white')

# Set text and axes color to white
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Set x-axis labels
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45)

# Set title and labels
ax.set_title('Sources', color='white')
ax.set_xlabel('')
ax.set_ylabel('')

# Add legend
ax.legend(loc='upper right', facecolor='black', edgecolor='white')

# Show plot
plt.show()
```
